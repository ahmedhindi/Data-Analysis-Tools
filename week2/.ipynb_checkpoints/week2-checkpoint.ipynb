# Chi-squared test

```python
import pandas as pd
import numpy as np
import scipy.stats
import itertools as itools
```

```python
#reading the data 
data = pd.read_pickle('data.pickle')
```

```python
data.head()
```

```python
data.ETHNICITY.value_counts()
```

```python
 list_of_vals = list(set((data.ETHNICITY.values)))
```

### assignment will use Chi-squared test to study the statistical significance of the relationship between ethnicity and being married or devoiced

```python
a = list(itools.combinations(list_of_vals,2))
print(a)
```

```python
sub3 = data[(data['marital'] == 'Divorced') | (data['marital'] =='Married')]
sub3 = sub3[['marital','ETHNICITY']].dropna()
sub3.head()
```

```python
for i in a:
    the_null = ' the null hypothesis "there is no relationship between being {} or {} and being Married or devoiced"'.format(i[0],i[1])
    remap = {i[0]:i[0],i[1]:i[1]}
    sub3['new_eth']= sub3['ETHNICITY'].map(remap)
    # contingency table of observed counts
    ct2=pd.crosstab(sub3['marital'], sub3['new_eth'])
    print(ct2)
    # column percentages
    colsum=ct2.sum(axis=0)
    colpct=ct2/colsum
    print(colpct)
    #hi-square
    print ('chi-square value    , p value    , expected counts')
    cs2= scipy.stats.chi2_contingency(ct2)
    print (cs2)
    print('_______________________________________________________________')
    if scipy.stats.chi2_contingency(ct2)[1] <= 0.003:
        print (the_null)
        print ('****** reject the null hypothesis  ')
    else:
        print (the_null)
        print ('****** accept the null hypothesis ')
    print('_______________________________________________________________')
    print('_______________________________________________________________')
```
